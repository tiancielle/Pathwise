```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python

# 1. Chargement des variables d'environnement (la clé API cachée)
load_dotenv()
```

```python
# Vérification de sécurité
if "GITHUB_API_KEY" not in os.environ:
    raise ValueError("La clé GITHUB_API_KEY est introuvable. Vérifiez votre fichier .env")
```

```python

# 2. Configuration du client
# L'astuce architecturale : on utilise la librairie OpenAI, mais on la redirige vers GitHub
client = OpenAI(
    base_url="https://models.inference.ai.azure.com",
    api_key=os.environ["GITHUB_API_KEY"],
)
```

```python
# 3. Appel au modèle gpt-4o-mini
print("Envoi de la requête au modèle en cours...")
try:
    response = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": "Tu es un professeur d'informatique. Sois clair, précis et concis.",
            },
            {
                "role": "user",
                "content": "Explique ce qu'est une API en une seule phrase simple.",
            }
        ],
        model="gpt-4o-mini",
        temperature=0.7,
    )

    # 4. Affichage de la réponse
    print("\n--- Réponse de GPT-4o-mini ---")
    print(response.choices[0].message.content)
except Exception as e:
    print(f"\nUne erreur est survenue : {e}")
```
